```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_distance(point1, point2):
    """
    Calculate the Euclidean distance between two points.

    Args:
        point1 (ndarray): First data point.
        point2 (ndarray): Second data point.

    Returns:
        float: Euclidean distance between the two points.
    """
    return np.sqrt(np.sum((point1 - point2)**2))

def single_linkage_distance(cluster1, cluster2):
    """
    Calculate the minimum distance between two clusters (single linkage).

    Args:
        cluster1 (list): List of data points in the first cluster.
        cluster2 (list): List of data points in the second cluster.

    Returns:
        float: Minimum distance between the two clusters.
    """
    min_distance = float('inf')
    for point1 in cluster1:
        for point2 in cluster2:
            distance = calculate_distance(point1, point2)
            if distance < min_distance:
                min_distance = distance
    return min_distance

def hierarchical_clustering(data):
    """
    Perform hierarchical clustering using the single linkage method.

    Args:
        data (ndarray): Array of data points.

    Returns:
        list: List containing the dendrogram information.
    """
    num_points = len(data)
    clusters = [[i] for i in range(num_points)]
    dendrogram = []

    while len(clusters) > 1:
        min_distance = float('inf')
        merge_indices = None

        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                distance = single_linkage_distance(clusters[i], clusters[j])
                if distance < min_distance:
                    min_distance = distance
                    merge_indices = (i, j)

        dendrogram.append([merge_indices[0], merge_indices[1], min_distance, len(clusters[merge_indices[0]]) + len(clusters[merge_indices[1]])])
        clusters.append(clusters[merge_indices[0]] + clusters[merge_indices[1]])
        del clusters[max(merge_indices)]
        del clusters[min(merge_indices)]

    return dendrogram

def plot_dendrogram(dendrogram):
    """
    Plot the dendrogram.

    Args:
        dendrogram (list): List containing the dendrogram information.
    """
    plt.figure(figsize=(10, 7))
    plt.title('Hierarchical Clustering Dendrogram')
    plt.xlabel('Data Point Index')
    plt.ylabel('Distance')
    for i, (idx1, idx2, distance, _) in enumerate(dendrogram):
        plt.plot([idx1, idx1], [0, distance], 'k')
        plt.plot([idx2, idx2], [0, distance], 'k')
        plt.plot([idx1, idx2], [distance, distance], 'k')
        plt.scatter([idx1, idx2], [distance, distance], color='k', marker='o')
    plt.show()

# Sample data points
data_points = np.array([[1000], [1500], [750], [2500], [1750], [1200], [1100], [1650], [3000], [1000]])

# Perform hierarchical clustering
dendrogram_info = hierarchical_clustering(data_points)

# Plot dendrogram
plot_dendrogram(dendrogram_info)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data points
data_points = np.array([[1000], [1500], [750], [2500], [1750], [1200], [1100], [1650], [3000], [1000]])

# Perform hierarchical clustering
linked = linkage(data_points, 'single')

# Plot dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Point Index')
plt.ylabel('Distance')
plt.show()
```
